```python
import pandas as pd
import numpy as np
from bs4 import BeautifulSoup
from lxml import etree
import requests
```

### HTTP

```python
url = 'https://www.marketwatch.com/investing/stock/aapl?mod=search_symbol'
response = requests.get(url)
```

### Soup and DOM

```python
soup = BeautifulSoup(response.content, 'html.parser')
dom = etree.HTML(str(soup))
soup
```

### Results

```python
open_value = dom.xpath('/html/body/div[3]/div[2]/div[3]/div/div[2]/h2/bg-quote')[0].text
open_value
```

```python
close_value = soup.find(class_='table__cell u-semi').text
close_value          
```

```python
day_range_from = dom.xpath('/html/body/div[3]/div[2]/div[5]/mw-rangebar[2]/div[1]/span[1]')[0].text
day_range_from
```

```python
day_range_to = dom.xpath('/html/body/div[3]/div[2]/div[5]/mw-rangebar[2]/div[1]/span[3]')[0].text
day_range_to
```

```python
week_52_range_from = soup.find(class_ = 'element element--range range--yearly').find(class_ = 'primary').text
week_52_range_from
```

```python
week_52_range_to = soup.find(class_ = 'element element--range range--yearly').findAll(class_ = 'primary')[1                                                                                                     ]
week_52_range_to = week_52_range_to.text
week_52_range_to
```

```python
analyst_ratings_list = soup.findAll(class_ = 'analyst__option')
analyst_ratings_list
```

```python
text = np.nan
for element in analyst_ratings_list:
    string_element = str(element)
    if 'active' in string_element:
        text = element.text
text
```

```python
data = {'Open value': open_value, 'Close value': close_value, 'Day range from': day_range_from, 
      'Day range to' : day_range_to, '52 Week range from': week_52_range_from,
      '52 Week range to': week_52_range_to, 'Analyst rating': text}
```

```python
df = pd.DataFrame.from_dict(data, orient = 'index')
df
```
